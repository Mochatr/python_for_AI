```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 600)
```

```python
activations = {
    "Relu": np.maximum(0, x),
    "Sigmoid": 1 / (1 + np.exp(-x)),
    "Tanh": np.tanh(x),
    "LeakyRelu": np.where(x > 0, x, 0.1 * x)
}
```

```python
# Dictionnaire des couleurs
couleurs = {
    "Relu": "tab:blue",
    "Sigmoid": "tab:orange",
    "Tanh": "tab:green",
    "LeakyRelu": "tab:red"
}

# Dictionnaire des marqueurs
marques = {
    "Relu": "o",       # cercle
    "Sigmoid": "s",    # carré
    "Tanh": "^",       # triangle haut
    "LeakyRelu": "D"   # losange
}
```

```python
# Création d'une nouvelle figure avec une taille adaptée
plt.figure(figsize=(10, 6))

# Parcours simultané des clés (noms) et valeurs (y) de nos activations
for nom, y in activations.items():
    # 
    # Afficher qu'un marqueur tous les 60 points
    plt.plot(x, y, 
             label=nom, 
             color=couleurs[nom], 
             marker=marques[nom], 
             markersize=6, 
             markevery=60)

# Ajout des éléments d'habillage et de lisibilité
plt.title("Comparaison des fonctions d'activation")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()

plt.savefig("activations_all.png", dpi=300)
plt.show()
```

```python
# Création d'une grille 2x2 avec plt.subplots 
fig, axes = plt.subplots(2, 2, figsize=(12, 8), sharex=True, sharey=True)

axes = axes.ravel()

for i, (nom, y) in enumerate(activations.items()):
    ax = axes[i]
    
    # Traçage d'activation par subplot avec ax.plot(...)
    ax.plot(x, y, 
            label=nom, 
            color=couleurs[nom], 
            marker=marques[nom], 
            markersize=6, 
            markevery=60)
    
    # Ajouter titre, labels, grille, légende dans chaque axe
    ax.set_title(f"Activation : {nom}")
    ax.set_xlabel("x") 
    ax.set_ylabel("f(x)")
    ax.grid(True) 
    ax.legend()

# éviter le chevauchement des textes et ajuster la figure [cite: 128, 42]
plt.tight_layout()

plt.savefig("activations_subplots.png", dpi=300)
plt.show()
```

```python
# Ajout des des fonctions supplémentaires
activations["Softplus"] = np.log(1 + np.exp(x))
activations["ELU"] = np.where(x > 0, x, 1.0 * (np.exp(x) - 1))

# Ajout des nouvelles couleurs
couleurs["Softplus"] = "tab:purple"
couleurs["ELU"] = "tab:brown"

# Ajout des nouveaux marqueurs
marques["Softplus"] = "*"  # étoile
marques["ELU"] = "v"       # triangle bas
```

```python
plt.figure(figsize=(10, 6))

for nom, y in activations.items():
    plt.plot(x, y, 
             label=nom, 
             color=couleurs[nom], 
             marker=marques[nom], 
             markersize=6, 
             markevery=60)

plt.title("Comparaison globale des fonctions d'activation (avec Extensions)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()

# Sauvegarde de la figure avec une résolution de 300 dpi
plt.savefig("activations_all_extension.png", dpi=300)
plt.show()
```

```python
# Grille de 2 lignes et 3 colonnes pour accueillir les 6 fonctions
fig, axes = plt.subplots(2, 3, figsize=(15, 10), sharex=True, sharey=True)
axes = axes.ravel()

for i, (nom, y) in enumerate(activations.items()):
    ax = axes[i]
    ax.plot(x, y, 
            label=nom, 
            color=couleurs[nom], 
            marker=marques[nom], 
            markersize=6, 
            markevery=60)
    
    ax.set_title(f"Activation : {nom}")
    ax.set_xlabel("x")
    ax.set_ylabel("f(x)")
    ax.grid(True)
    ax.legend()

plt.tight_layout()

# Sauvegarde de la figure avec une résolution de 300 dpi
plt.savefig("activations_subplots_extension.png", dpi=300)
plt.show()
```
